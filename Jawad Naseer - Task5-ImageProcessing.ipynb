# Important Function
# OpenCV (`cv2`) Functions Overview


## 1. `cv2.imread()`
**Purpose**: Load an image from a file.

**Parameters**:
- `filename`: Path to the image file.
- `flags`: Image loading flags (e.g., `cv2.IMREAD_GRAYSCALE`, `cv2.IMREAD_COLOR`).

**Example**:
```python
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
```
##  2. `cv2.resize()`

**Purpose:** Resize an image.

Parameters:

- **src:** Input image.
- **dsize:** Output size as (width, height).
- **fx:** Scale factor along the horizontal axis (optional).
- **fy:** Scale factor along the vertical axis (optional).
- **interpolation:** Interpolation method (e.g., cv2.INTER_LINEAR).

```python
resized_image = cv2.resize('image.jpg', (width, height), interpolation=cv2.INTER_LINEAR)
```

## 3. `cv2.GaussianBlur()`

**Purpose**: Apply Gaussian blur to an image to reduce noise and detail. This is often used for smoothing images.

**Description**: The `cv2.GaussianBlur()` function applies a Gaussian filter to the image, which blurs the image and reduces its details. This can be useful for noise reduction or smoothing before applying other image processing techniques.

**Parameters**:
- **`src`**: Input image.
  - **Type**: `numpy.ndarray`
- **`ksize`**: Kernel size as `(width, height)`. Both dimensions must be odd numbers. The size of the kernel affects the extent of the blur.
  - **Type**: `tuple`
  - **Example**: `(5, 5)`
- **`sigmaX`**: Standard deviation in the X direction. A larger value results in a more pronounced blur.
  - **Type**: `float`
  - **Example**: `0`

**Example**:
```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```
## 4.  `cv2.filter2D()`

**Purpose**: Apply a custom convolutional filter to an image. This allows for advanced filtering operations like sharpening or edge detection.

**Description**: The `cv2.filter2D()` function convolves an image with a kernel. This kernel can be used to perform various image processing tasks such as sharpening, blurring, or edge detection based on the kernel values.

**Parameters**:
- **`src`**: Input image.
  - **Type**: `numpy.ndarray`
- **`ddepth`**: Depth of the output image. Typically set to `-1` to use the same depth as the source image.
  - **Type**: `int`
  - **Example**: `-1`
- **`kernel`**: Convolution kernel. A matrix that defines the filter.
  - **Type**: `numpy.ndarray`

**Example**:
```python
filtered_image = cv2.filter2D(image, -1, kernel)
```
## 5.  `cv2.threshold()`

**Purpose**: Apply a fixed-level threshold to an image, which is useful for image segmentation and binarization.

**Description**: The `cv2.threshold()` function converts an image to a binary image by applying a threshold. Pixels with values above the threshold are set to a maximum value, while others are set to 0 or another specified value. This is commonly used for image segmentation tasks.

**Parameters**:
- **`src`**: Input image.
  - **Type**: `numpy.ndarray`
- **`thresh`**: Threshold value. Pixels with a value greater than this will be set to `maxval`.
  - **Type**: `float`
- **`maxval`**: Maximum value to use with the `THRESH_BINARY` and `THRESH_BINARY_INV` thresholding types.
  - **Type**: `float`
- **`type`**: Thresholding type. Examples include `cv2.THRESH_BINARY` for binary thresholding or `cv2.THRESH_BINARY_INV` for inverted binary thresholding.
  - **Type**: `int`
  - **Example**: `cv2.THRESH_BINARY`

**Example**:
```python
thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```
## 6.  `cv2.Canny()`

**Purpose**: Apply the Canny edge detection algorithm to identify edges in an image. This is useful for detecting boundaries and features.

**Description**: The `cv2.Canny()` function performs edge detection using the Canny algorithm. This method identifies edges by detecting areas of the image with a high gradient, which correspond to boundaries between different regions.

**Parameters**:
- **`image`**: Input image.
  - **Type**: `numpy.ndarray`
- **`threshold1`**: Lower threshold for the hysteresis procedure. Edge pixels with gradient values higher than `threshold2` are considered strong edges.
  - **Type**: `float`
- **`threshold2`**: Upper threshold for the hysteresis procedure. Pixels with gradient values between `threshold1` and `threshold2` are considered weak edges.
  - **Type**: `float`

**Example**:
```python
edges = cv2.Canny(image, 100, 200)
```

## 7.   `cv2.HoughCircles()`

**Purpose**: Detect circles in an image using the Hough Circle Transform. This is useful for finding circular shapes in images.

**Description**: The `cv2.HoughCircles()` function detects circles in an image by using the Hough Circle Transform. It identifies circles based on the parameters provided, such as the minimum distance between circles and the accumulator thresholds.

**Parameters**:
- **`image`**: Input image.
  - **Type**: `numpy.ndarray`
- **`method`**: Detection method. Typically `cv2.HOUGH_GRADIENT`.
  - **Type**: `int`
- **`dp`**: Inverse ratio of the accumulator resolution to the image resolution. A value of 1 means the accumulator has the same resolution as the image.
  - **Type**: `float`
- **`minDist`**: Minimum distance between the centers of the detected circles.
  - **Type**: `float`
- **`param1`**: First method-specific parameter, such as the higher threshold for the Canny edge detector.
  - **Type**: `float`
- **`param2`**: Second method-specific parameter, such as the accumulator threshold for the circle centers.
  - **Type**: `float`

**Example**:
```python
circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=30)
```

# Convolution
**Convolution** is a mathematical operation used to blend two functions. In the context of image processing, convolution is applied to an image with a filter (or kernel) to extract features such as edges, blur, or sharpen the image.

The process involves moving the kernel over the image, pixel by pixel, and at each position, performing an element-wise multiplication between the kernel values and the corresponding image pixel values. The sum of these multiplications gives the output pixel value for that position.

## Example in Image Processing:
For example, a **3x3** kernel can be used to detect edges in an image. The kernel slides over the image, multiplying its values with the corresponding pixel values and then summing the result to produce a single pixel in the output image.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/sample_data/parrot.jpg', 0)

# Define an edge detection kernel (Sobel operator)
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# Apply the convolution using OpenCV's filter2D function
convoluted_image = cv2.filter2D(image, -1, kernel)

# Use matplotlib to display the original and convoluted images
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display edge-detected image (convoluted image)
plt.subplot(1, 2, 2)
plt.imshow(convoluted_image, cmap='gray')
plt.title('Edge Detected Image')
plt.axis('off')

# Show both images
plt.tight_layout()
plt.show()

````

# Gray Scalling
Gray-scaling is a technique in image processing where a color image is converted into a grayscale (or black-and-white) image. This process reduces the complexity of an image by removing color information, leaving only the intensity of light. Each pixel in a grayscale image represents a shade of gray, ranging from 0 (black) to 255 (white).

Why Gray-scaling?
Simplification: Gray-scaling reduces the complexity of the image, making it easier to process for tasks like edge detection or object recognition.
Efficiency: Processing grayscale images is faster as they require less computational resources compared to colored images.
Preprocessing Step: Many machine learning models and computer vision algorithms work better with grayscale images.

````python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1, 2, figsize=(16, 8))
fig.tight_layout()

ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original")

ax[1].imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
ax[1].set_title("Grayscale")
plt.show()
````

# Task 12 Linear Spatial Filtering

**Description:**  
Linear spatial filtering is a technique used in image processing to modify an image based on the linear combination of pixel values in its neighborhood. This process involves using a filter (or kernel) that slides over the image to perform operations such as smoothing, sharpening, and edge detection.

---

### **Key Concepts:**

1. **Filter (Kernel):**  
   - A small matrix (e.g., 3x3, 5x5) used to process the image. Each element in the kernel represents a weight that modifies the corresponding pixel in the image.

2. **Convolution:**  
   - The core operation in linear spatial filtering. It involves overlaying the filter on the image, multiplying the filter values with the corresponding pixel values, summing the results, and replacing the central pixel with this sum.

3. **Filtering Operations:**  
   - **Smoothing (Blurring):**  
     Reduces noise and detail by averaging pixel values, which results in a softening effect. Common kernels include the average filter and Gaussian filter.
   - **Sharpening:**  
     Enhances edges and fine details by emphasizing differences between pixel values. Common kernels include the Laplacian filter and the unsharp mask.
   - **Edge Detection:**  
     Highlights the boundaries of objects within an image. Filters like the Sobel and Prewitt operators are commonly used for this purpose.
